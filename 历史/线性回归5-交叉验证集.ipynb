```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import sklearn
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold

from sklearn.linear_model import BayesianRidge, LinearRegression, ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_excel('所有变量.xls')
df = DataFrame(data)
df.head
```

```python
X = df.iloc[:,3:]
y = df.iloc[:,2:3]

X_train,X_test, y_train,y_test = train_test_split(X, y, 
                                                 test_size=0.3,  # 划分比例
                                                 random_state=0)   # 随机数种子
X_train
```

```python
#归一化
scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)
```

```python
#降维
pca = PCA(n_components = 60)
pca.fit(X_train)
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

```python
kf = KFold(n_splits=5, shuffle=True)
for train_index, test_index in kf.split(X_train):
    X_test
```

```python
model_br = BayesianRidge()  # 建立贝叶斯回归模型

model_lr = LinearRegression() # 建立普通线性回归模型

model_etc = ElasticNet() # 建立弹性网络回归模型

model_svr = SVR() # 建立支持向量回归模型

model_gbr = GradientBoostingRegressor() # 建立梯度增强回归模型

model_names = ['BayesianRidge','LinearRegression','ElasticNet', 'SVR', 'GBR']

model_dir = [model_br, model_lr, model_etc, model_svr, model_gbr]
```

```python
# 交叉验证评分与模型训练
n_folds = 5

cv_score_list = [] # 交叉检验结果列表

y_train_pre = [] # 各个模型预测的y值列表

y_test_pre = [] # 创建测试集预测结果列表
```

```python
for model in model_dir:
    scores = cross_val_score(model, X_train, np.ravel(y_train), cv = n_folds, scoring = 'r2')  #对每个回归模型进行交叉验证
    cv_score_list.append(scores) # 将验证结果保存在列表中
    
    #将训练模型的预测结果保存在列表中
    y_train_pre.append(model.fit(X_train, np.ravel(y_train)).predict(X_train))
    y_test_pre.append(model.fit(X_train, np.ravel(y_train)).predict(X_test))
```
